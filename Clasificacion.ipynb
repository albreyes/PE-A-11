```python
import pandas as pd
data_path = 'C:/Users/areyes/Dropbox/ineel/11090/pronostico/datasets/tutorialenc/titanic.csv'
titanic=pd.read_csv(data_path,encoding="latin",low_memory=False)
print(titanic.shape)
titanic.info()
```

```python
def simplify_ages(df):
    df.Age = df.Age.fillna(-0.5)
    bins = (-1, 0, 5, 12, 18, 25, 35, 60, 120)
    group_names = ['Unknown', 'Baby', 'Child', 'Teenager', 'Student','Young Adult', 'Adult', 'Senior']
    categories = pd.cut(df.Age, bins, labels=group_names)
    # La función CUT de Pandas nos regresa el índice 
    # del bin al que pertenece el valor x
    df.Age = categories
    return df
```

```python
def simplify_cabins(df):
    df.Cabin = df.Cabin.fillna('N')
    df.Cabin = df.Cabin.apply(lambda xx: xx[0])
    return df
```

```python
def simplify_fares(df):
    df.Fare = df.Fare.fillna(-0.5)
    bins = (-1, 0, 8, 15, 31, 1000) #titanic_train.Fare.describe()
    group_names = ['Unknown', '1_quartile', '2_quartile', '3_quartile', '4_quartile']
    categories = pd.cut(df.Fare, bins, labels=group_names)
    # Lo mismo que vimos anteriormente con la función CUT
    df.Fare = categories
    return df
```

```python
def format_name(df):
    df['Lname'] = df.Name.apply(lambda x: x.split(' ')[0])
    df['NamePrefix'] = df.Name.apply(lambda x: x.split(' ')[1])
    return df
```

```python
def drop_features(df):
    return df.drop(['Ticket', 'Name', 'Embarked'], axis=1)
```

```python
def transform_features(df):
    df = simplify_ages(df)
    df = simplify_cabins(df)
    df = simplify_fares(df)
    df = format_name(df)
    df = drop_features(df)
    return df
```

```python
transform_features(titanic)
```

```python
from sklearn import preprocessing
def encode_features(datos):
    features = ['Fare', 'Cabin', 'Age', 'Sex', 'Lname', 'NamePrefix']
    #df_combined = pd.concat([df_train[features], df_test[features]])
    
    for feature in features:
        le = preprocessing.LabelEncoder()
        le = le.fit(datos[feature])
        datos[feature] = le.transform(datos[feature])
        
    return datos
```

```python
d_encode = encode_features(titanic)
d_encode.head()
```

```python
from sklearn.model_selection import train_test_split
X_all = d_encode.drop(['Survived', 'PassengerId', 'Name', 'Ticket', 'Embarked'], axis=1)

#Borramos las columnas Survived 
y_all = d_encode['Survived']

num_test = 0.20 # aquí se está especificando 20% para test
X_train, X_test, y_train, y_test = train_test_split(X_all, y_all, test_size=num_test)
```

```python
# Seleccionamos el tipo de clasificador, en este caso RandomForest. 
from sklearn.ensemble import RandomForestClassifier
clf = RandomForestClassifier(n_estimators=10)
# entrenamiento
clf.fit(X_train, y_train)
```

```python
from sklearn.metrics import make_scorer, accuracy_score, f1_score
predictions = clf.predict(X_test)
# se imprime el accuracy
print("Accuracy: ", accuracy_score(y_test, predictions))
#Imprimimos la f-score
print("F1-Score: ", f1_score(y_test, predictions))
```

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(clf, X_train, y_train, cv=2)
print(scores)
```
